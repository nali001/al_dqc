```python
import os
import glob
import math
import pandas as pd
import matplotlib.pyplot as plt


def group_files_by_conditions(directory, comparison_pairs, key_variables):
    grouped_files = {}
    for n_initial, k, initial_method, model_name, query_strategy in comparison_pairs: 
    # Groups the files based on their n_initial and k values
        files = glob.glob(os.path.join(directory, query_strategy, f'{model_name}_{initial_method}_{n_initial}_initial_{k}_k_*_rand.csv'))
        key = []
        for v in key_variables: 
            key.append(eval(v))
        key = tuple(key)

        if key not in grouped_files:
            grouped_files[key]=files
    
    return grouped_files


# Function to read CSV files and calculate mean and standard deviation
def calculate_mean_std(file_paths, metric):
    data_frames = []
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        data_frames.append(df[metric].values)
    # Concatenate data frames vertically (assuming they have the same structure)
    combined_df = pd.DataFrame(data_frames)
    # Calculate mean and standard deviation
    mean_values = combined_df.mean(axis=0).values
    std_values = combined_df.std(axis=0).values / math.sqrt(len(file_paths))
    return mean_values, std_values            


def read_num_samples(file_path):
    # Reads the number of samples from a CSV file
    df = pd.read_csv(file_path)
    return df['num_samples'].values # Assuming num_samples is in the first column

```

```python
def find_first_index_with_differences(numbers, diff_threshold):
    """
    Finds the index of the element that has first reached the highest number,
    with differences greater than or equal to the threshold for a consecutive count.

    Parameters:
    - numbers: List of numbers representing the sequence.
    - diff_threshold: The minimum difference between consecutive numbers to consider.
    - consecutive_count: The number of consecutive numbers with differences greater
      than or equal to the threshold to be considered.

    Returns:
    - The index of the element that has first reached the highest number.
    - Returns -1 if no such index is found.
    """
    highest_num = 0
    highest_index = -1
    
    for i, num in enumerate(numbers):
        if num - diff_threshold> highest_num:
            highest_num = num
            highest_index = i
        else:
            continue
    return highest_index  # If no such index is found
```

```python
def plot_metrics(grouped_files, metric):
    plt.figure(figsize=[5,5])

    # Plots the F1 scores for different models using the number of samples as the x-axis
    for key, files in grouped_files.items():
        print(key)
        mean_values, std_values = calculate_mean_std(files, metric)

        num_samples = read_num_samples(files[0])  # Assuming all files have the same num_samples
        
        # Find the point where the model becomes stable 
        stable_index = find_first_index_with_differences(mean_values, 0.01)
        print(f'{metric}: {mean_values[-1].round(4)} Stable point: {num_samples[stable_index]}; {mean_values[stable_index].round(4)}')
        
        plt.plot(num_samples, mean_values, label=key[0], linewidth=1.5, color=COLOR_MAP[key[0]], linestyle = LINE_MAP[key[0]])
        plt.fill_between(num_samples, mean_values - std_values, mean_values + std_values, color=COLOR_MAP[key[0]], alpha=0.1)
        # plt.vlines(num_samples, mean_values - std_values, mean_values + std_values, color=COLOR_MAP[key[0]], linestyle='dashed', linewidth=2)
        # plt.plot(num_samples, value, label=key, linewidth=1.5, color=COLOR_MAP[key[1]], linestyle = LINE_MAP[key[0]], marker=MARKER_MAP[key[0]], markerfacecolor='none', markersize=5)

    plt.xlabel('# queried samples')
    plt.ylabel(metric)
    plt.ylim([-0.05, 1.05])
    # plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.4), ncol=2, fancybox=True)
    plt.legend(loc='lower right')
    # plt.grid(True)
    # print(f'========== Initial:{n_initial}   K:{k} ==========')
    
    figure_file = os.path.join(PLOT_PATH, f'{metric}_{float_number}_{initial_method}_{key[1]}_rand.png')
    plt.savefig(figure_file, dpi=300, bbox_inches='tight')
    plt.show()

# Example usage
print(f'------ {float_number} -------')
directory = RESULT_PATH  # Replace with the actual directory path
grouped_files = group_files_by_conditions(directory, comparison_pairs, key_variables)
plot_metrics(grouped_files, 'F1-score')
plot_metrics(grouped_files, 'Precision')
plot_metrics(grouped_files, 'Recall')
```
